```julia
function step(template)
    after = template[1]
    for i in 2:length(template)
        after *= rules[template[(i-1):i]] * template[i]
    end
    return after
end
```

### Alternative approach

iteratively count each letter after a certain number of steps

Rules contain all possible pairs.
So for each pair calculate the number of each letter after 10 steps

```julia
using Lazy
```

```julia
itstep = iterated(step, template)
```

```julia
itstep[10] |> collect |> countmap |> sort |> values |> collect
```

```julia
letter_counts_10 = Dict()
for pair in keys(rules)
    letter_counts_10[pair] = iterated(step, pair)[11] |> collect |> countmap |> sort |> values |> collect # 11 because element 1 is x not f(x)
end
letter_counts_10
```

```julia
# check letter completeness
unique([length(x) for x in values(letter_counts_10)])
```

As an exercise check part I with this approach ðŸ˜‚

```julia
all_letters = first.(sort(unique(values(rules)))) # first is to convert single character String to Char
```

```julia
# helper function to get a count vector from a string with (potentially) missing letters
function manual_letter_count(seq)
    counts = zeros(Int, 10)
    cm = seq |> collect |> countmap
    for i in 1:10
        if haskey(cm, all_letters[i])
            counts[i] = cm[all_letters[i]]
        end
    end
    return counts
end
```

```julia
manual_letter_count(template)
```

```julia
function total_letter_counts(seq, counts)
    total = zeros(Int, 10)
    for i in 2:length(seq)
        total .+= counts[seq[(i-1):i]]
    end
    # now the overlapping characters are counted twice
    total .-= manual_letter_count(seq[2:(end-1)])
    return total
end
```

```julia
tlc = total_letter_counts(template, letter_counts_10)
maximum(tlc) - minimum(tlc)
```

Perfect! So now iterate the counting for 20, 30 and 40 steps

```julia
total_letter_counts(iterated(step, "PK")[11], letter_counts_10)
```

```julia
letter_counts_20 = Dict()
for pair in keys(rules)
    letter_counts_20[pair] = total_letter_counts(iterated(step, pair)[11], letter_counts_10)
end
letter_counts_20
```

```julia
letter_counts_30 = Dict()
for pair in keys(rules)
    letter_counts_30[pair] = total_letter_counts(iterated(step, pair)[11], letter_counts_20)
end
letter_counts_30
```

```julia
letter_counts_40 = Dict()
for pair in keys(rules)
    letter_counts_40[pair] = total_letter_counts(iterated(step, pair)[11], letter_counts_30)
end
letter_counts_40
```

```julia
tlc = total_letter_counts(template, letter_counts_40)
maximum(tlc) - minimum(tlc)
```
