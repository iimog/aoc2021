```julia
byte2bit = Dict(
    '0' => "0000",
    '1' => "0001",
    '2' => "0010",
    '3' => "0011",
    '4' => "0100",
    '5' => "0101",
    '6' => "0110",
    '7' => "0111",
    '8' => "1000",
    '9' => "1001",
    'A' => "1010",
    'B' => "1011",
    'C' => "1100",
    'D' => "1101",
    'E' => "1110",
    'F' => "1111"
);
```

```julia
bits = join([byte2bit[x] for x in collect(bytes)])
```

```julia
function parse_package(s)
    version = parse(Int, "0b"*s[1:3])
    typeId = parse(Int, "0b"*s[4:6])
    subpack = []
    if typeId == 4
        length = 7
        while(s[length] != '0')
            length += 5
        end
        length += 4
    else
        lenType = s[7]
        if lenType == '0'
            bitLength = parse(Int, "0b"*s[8:22])
            length = bitLength + 22
            pos = 23
            while(pos<length)
                (v,t,l,p) = parse_package(s[pos:end])
                push!(subpack, (v,t,l,p))
                pos += l
            end
        else
            chunkNum = parse(Int, "0b"*s[8:18])
            pos = 19
            for i in 1:chunkNum
                (v,t,l,p) = parse_package(s[pos:end])
                push!(subpack,(v,t,l,p))
                pos += l
            end
            length = pos-1
        end
    end
    return version, typeId, length, subpack
end
```

```julia
packs = parse_package(bits)
```

```julia
function version_sum_pack(pack)
    vs = pack[1]
    for sp in pack[4]
        vs += version_sum_pack(sp)
    end
    return vs
end
```

```julia
function parse_package_with_value(s)
    version = parse(Int, "0b"*s[1:3])
    typeId = parse(Int, "0b"*s[4:6])
    subpack = []
    value = -1
    if typeId == 4
        length = 7
        valuebits = "0b"*s[(length+1):(length+4)]
        while(s[length] != '0')
            length += 5
            valuebits *= s[(length+1):(length+4)]
        end
        length += 4
        value = parse(Int, valuebits)
    else
        lenType = s[7]
        if lenType == '0'
            bitLength = parse(Int, "0b"*s[8:22])
            length = bitLength + 22
            pos = 23
            while(pos<length)
                (v,t,l,p,val) = parse_package_with_value(s[pos:end])
                push!(subpack, (v,t,l,p,val))
                pos += l
            end
        else
            chunkNum = parse(Int, "0b"*s[8:18])
            pos = 19
            for i in 1:chunkNum
                (v,t,l,p,val) = parse_package_with_value(s[pos:end])
                push!(subpack,(v,t,l,p,val))
                pos += l
            end
            length = pos-1
        end
    end
    return version, typeId, length, subpack, value
end
```

```julia
pack_with_value = parse_package_with_value(bits)
```

```julia
function get_value(pack)
    value = pack[5]
    if value<0
        subvalues = [get_value(sp) for sp in pack[4]]
        if pack[2] == 0
            value = sum(subvalues)
        elseif pack[2] == 1
            value = prod(subvalues)
        elseif pack[2] == 2
            value = minimum(subvalues)
        elseif pack[2] == 3
            value = maximum(subvalues)
        elseif pack[2] == 5
            value = subvalues[1] > subvalues[2] ? 1 : 0
        elseif pack[2] == 6
            value = subvalues[1] < subvalues[2] ? 1 : 0
        elseif pack[2] == 7
            value = subvalues[1] == subvalues[2] ? 1 : 0
        end
    end
    return value
end
```
