```julia
scanners = filter(x->x!="", lines) |> x->join(x,';') |> x->split(x,r"--- scanner \d+ ---;") |> x->x[2:end]
```

```julia
parseScanner(x) = split(x, ";") |> z->z[1:(end-1)] |> z->split.(z,",") |> z->reduce(hcat,z) |> z->parse.(Int,z) |> transpose |> collect
```

```julia
scanners = parseScanner.(scanners)
```

```julia
function all_orientations(s)
    orient = [s, hcat(s[:,2],-s[:,1],s[:,3]), hcat(s[:,3],s[:,2],-s[:,1])]
    for i in 1:3
        c = orient[i]
        push!(orient, hcat(-c[:,1],-c[:,2],c[:,3]))
    end
    for i in 1:6
        c = orient[i]
        push!(orient, hcat(c[:,1],-c[:,2],-c[:,3]))
        push!(orient, hcat(c[:,1],-c[:,3],c[:,2]))
        push!(orient, hcat(c[:,1],c[:,3],-c[:,2]))
    end
    return orient
end
```

```julia
function merge_scanners(s1, s2)
    s2o = all_orientations(s2)
    for cs2 in s2o
        for b1 in eachrow(s1)
            for b2 in eachrow(cs2)
                shift = b2 - b1
                shifted_cs2 = cs2 .- repeat(shift', size(cs2)[1])
                overlap = sum(collect(eachrow(s1)) .∈ [collect(eachrow(shifted_cs2))])
                if overlap >= 12 # limitations, this only checks if there are at least 12 overlapping beacons, it does not check whether there are any contradicting ones...
                    return unique(vcat(s1, shifted_cs2), dims=1)
                end
            end
        end
    end
    return false
end
```

```julia
merge_scanners(scanners[1], scanners[2])
```

```julia
full = scanners[1]
remaining_scanners = scanners[2:end]
while length(remaining_scanners)>0
    remaining_indices = []
    for (i,s) in enumerate(remaining_scanners)
        new_map = merge_scanners(full, s)
        if new_map isa Matrix
            full = new_map
        else
            push!(remaining_indices, i)
        end
    end
    remaining_scanners = remaining_scanners[remaining_indices]
    print(remaining_indices)
end
```

```julia
function merge_scanners_with_shift(s1, s2)
    s2o = all_orientations(s2)
    for cs2 in s2o
        for b1 in eachrow(s1)
            for b2 in eachrow(cs2)
                shift = b2 - b1
                shifted_cs2 = cs2 .- repeat(shift', size(cs2)[1])
                overlap = sum(collect(eachrow(s1)) .∈ [collect(eachrow(shifted_cs2))])
                if overlap >= 12
                    return unique(vcat(s1, shifted_cs2), dims=1), shift
                end
            end
        end
    end
    return false, false
end
```

```julia
full = scanners[1]
remaining_scanners = scanners[2:end]
scanner_pos = []
while length(remaining_scanners)>0
    remaining_indices = []
    for (i,s) in enumerate(remaining_scanners)
        new_map, sh = merge_scanners_with_shift(full, s)
        if new_map isa Matrix
            full = new_map
            push!(scanner_pos, sh)
        else
            push!(remaining_indices, i)
        end
    end
    remaining_scanners = remaining_scanners[remaining_indices]
    println(remaining_indices)
end
```
