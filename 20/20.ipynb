```julia
lines = readlines("input")
```

```julia
using Underscores
```

```julia
enhance = @_ lines[1] |> collect |> replace(__, '#'=>1, '.'=>0)
```

```julia
field = lines[3:end] .|> collect |> x->replace.(x, '#'=>1, '.'=>0) |> x->reduce(hcat, x) |> transpose
```

```julia
function pad_zeros(A,n)
    s = size(A)
    B = zeros(Int, s[1]+2*n, s[2]+2*n)
    B[n+1:n+s[1], n+1:n+s[2]] .= A
    return B
end
```

```julia
pad_zeros([1 2; 3 4], 3)
```

```julia
padded_field = pad_zeros(field, 5)
```

```julia
m = [1 2 3; 4 5 6; 7 8 9]
m'[7]
```

```julia
function pos2index(field, pos)
    area = field[pos[1]:pos[1]+2,pos[2]:pos[2]+2]
    ind = 0
    for i in 1:9
        ind += 2^(9-i)*area'[i]
    end
    return ind
end
```

```julia
pos2index(field, (1,1))
```

```julia
function step_field(field)
    new_field = zeros(Int, size(field)[1]-2, size(field)[2]-2)
    for i in 1:size(field)[1]-2
        for j in 1:size(field)[2]-2
            new_field[i,j] = enhance[pos2index(field, (i,j))+1]
        end
    end
    return new_field
end
```

```julia
after1 = step_field(padded_field)
```

```julia
after2 = step_field(after1)
```

```julia
sum(after2)
```
