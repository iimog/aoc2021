```julia
lines = readlines("input") .|> collect |> x->replace.(x, '.'=>0, '>'=>1, 'v'=>2)
seamap = reduce(vcat, map(x->x',lines))
```

```julia
shift_right(A) = hcat(A[:, end],A[:, 1:end-1])
shift_left(A) = hcat(A[:, 2:end],A[:, 1])
shift_up(A) = vcat(A[2:end, :],A[1:1, :])
shift_down(A) = vcat(A[end:end, :],A[1:end-1, :])
```

```julia
herd1_pos = seamap .== 1
```

```julia
free = seamap .== 0
```

```julia
herd1_shift = shift_right(herd1_pos) .& free
```

```julia
seamap[herd1_shift] .= 1
seamap[shift_left(herd1_shift)] .= 0;
```

```julia
function make_step(A)
    changes = 0
    herd1_pos = A .== 1
    free = A .== 0
    herd1_shift = shift_right(herd1_pos) .& free
    A[herd1_shift] .= 1
    A[shift_left(herd1_shift)] .= 0
    changes += sum(herd1_shift)
    herd2_pos = A .== 2
    free = A .== 0
    herd2_shift = shift_down(herd2_pos) .& free
    A[herd2_shift] .= 2
    A[shift_up(herd2_shift)] .= 0
    changes += sum(herd2_shift)
    return changes
end
```

```julia
make_step(seamap)
```

```julia
# re-initialize
lines = readlines("input") .|> collect |> x->replace.(x, '.'=>0, '>'=>1, 'v'=>2)
seamap = reduce(vcat, map(x->x',lines));
```

```julia
steps = 0
changes = 1
while changes > 0
    changes = make_step(seamap)
    steps += 1
end
steps
```
