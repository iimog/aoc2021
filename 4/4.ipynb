```julia
lines = readlines("input")
```

```julia
draws = [parse(Int, x) for x in split(lines[1], ",")];
```

```julia
function fresh_boards()
    boards = []
    for i in 2:6:length(lines)
        mat = zeros(Int,5,5)
        for j in 1:5
            row = [parse(Int,x) for x in split(lstrip(lines[i+j]), r"\s+")]
            mat[j,:] = row
        end
        push!(boards, mat)
    end
    return boards
end
```

```julia
boards = fresh_boards()
```

## Part I

Confirm that there are no negative numbers. Thus -1 can be used as mark and row/col sums of -5 mean bingo. Need to be set to 0 before calculating the score.

```julia
minimum([minimum(b) for b in boards])
```

```julia
function get_bingo()
    for i in 1:length(draws)
        current = draws[i]
        for b in boards
            b[b .== current] .= -1
            if -5 in sum(b,dims=1) || -5 in sum(b,dims=2)
                print(b)
                b[b .== -1] .= 0
                return sum(b)*current
            end
        end
    end
end

get_bingo()
```

```julia
function get_all_bingos()
    boards = fresh_boards()
    for i in 1:length(draws)
        current = draws[i]
        for b in boards
            b[b .== current] .= -1
            if -5 in sum(b,dims=1) || -5 in sum(b,dims=2)
                print(b,"\t")
                b[b .== -1] .= 0
                print(sum(b)*current,"\n")
                # invalidate board to avoid more bingos
                b .= -2
            end
        end
    end
end

get_all_bingos()
```
