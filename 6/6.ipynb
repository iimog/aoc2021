```julia
x = readline("input") |> x->parse.(Int,split(x,","))
```

```julia
using StatsBase
```

```julia
cm = countmap(x)
```

```julia
lanterns = zeros(Int,9);
for k in keys(cm)
    lanterns[k+1] = cm[k]
end
lanterns'
```

```julia
for i in 1:80
    lanterns = [lanterns[2:end]; lanterns[1]]
    lanterns[7] += lanterns[end]
    print(lanterns,"\n")
end
```

```julia
sum(lanterns)
```
